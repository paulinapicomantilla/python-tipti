```python
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt # data visualization

data_path = '../data/processed/cleaned_products.csv'#path to the data
df=pd.read_csv(data_path)#load the data


print("Basic statistics of the data")#print basic statistics of the data
print(df.describe())#print basic statistics of the data


plt.hist(df['price'], bins=100,edgecolor='black')#plot the histogram of the prices
plt.title('Distribucion de precios', color='orange')#title of the plot
plt.xlabel('Precio eje de las X', color='blue')#x axis label
plt.ylabel('Frecuencia eje de las Y', color='red')#y axis label
plt.show()#show the plot

print("productos mas caros")#print the most expensive products
highestPrice=df.nlargest(6, 'price')#get the 6 most expensive products
print(highestPrice)#print the 6 most expensive products
```
